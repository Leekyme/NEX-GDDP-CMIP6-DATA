```python
import matplotlib.pyplot as plt
import numpy as np
import geocat.viz as gv
```

```python
#Metrics
CA_ratio = [0.724, 0.482, 0.764, 0.712, 0.797, 0.570, 0.649, 0.752, 0.001, 0.614, 0.343, 0.751]  # Input your TSS Values
CA_cc = [0.792, 0.571, 0.781, 0.698, 0.786, 0.669, 0.759, 0.741, -0.013, 0.587, 0.772, 0.735]  # correlation coefficient
fig = plt.figure(figsize=(12, 12))
dia = gv.TaylorDiagram(fig=fig, label='REF')
ax = plt.gca()
dia.add_model_set(
    CA_ratio,
    CA_cc,
    fontsize=10,
    xytext=(-5, 10),  # marker label location, in pixels
    color = 'magenta', #you can play with the colors on how you want.
    marker= 'D', #you can also play with the marker part. All in all, expolore!
    s=20)  # marker size
dia.add_contours(levels=np.arange(0, 1.1, 0.25),
                 colors='lightgrey',
                 linewidths=0.5)
dia.add_std_grid(np.array([0.5, 1.5]), color='grey')
dia.add_corr_grid(np.array([0.6, 0.9]))
plt.title("OND Precipitation", size=20, pad=40)
plt.show()
```
